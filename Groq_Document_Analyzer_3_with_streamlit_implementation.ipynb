```python
%%writefile "app.py"
import streamlit as st
import requests
import base64
from PIL import Image
import json

# Groq API credentials and URL
GROQ_API_URL = "https://api.groq.com/openai/v1/chat/completions"
GROQ_API_KEY = "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"

def analyze_image_with_groq(image_path):
    """
    Analyze an image using the Groq API.
    """
    try:
        # Open image and encode it in base64
        with open(image_path, "rb") as image_file:
            image_data = image_file.read()
            base64_image = base64.b64encode(image_data).decode('utf-8')  # Encode to base64 and convert to string

        headers = {
            "Authorization": f"Bearer {GROQ_API_KEY}",
            "Content-Type": "application/json"
        }

        # Prepare the payload to send image data along with the model choice
        payload = {
            "model": "llama-3.2-90b-vision-preview",  # Model that can process images
            "messages": [
                {
                    "role": "user",
                    "content": [
                        {
                            "type": "text",
                            "text": "What's in this image?"  # Adjust this prompt as per need
                        },
                        {
                            "type": "image_url",
                            "image_url": {
                                "url": f"data:image/jpeg;base64,{base64_image}"  # Use base64-encoded image
                            }
                        }
                    ]
                }
            ],
            "temperature": 1,
            "max_completion_tokens": 1024,
            "top_p": 1,
            "stream": False
        }

        response = requests.post(GROQ_API_URL, headers=headers, json=payload)

        if response.status_code == 200:
            response_json = response.json()  # Parse the JSON response
            return response_json['choices'][0]['message']['content']  # Get content from JSON response
        else:
            st.error(f"Error from Groq API: {response.status_code} - {response.text}")
            return None
    except Exception as e:
        st.error(f"Error connecting to Groq API: {e}")
        return None

def parse_dna_analysis(content):
    """
    Parse and filter the key DNA analysis information from the content.
    """
    analysis_keywords = [
        "DNA paternity test", "alleged father", "child sample",
        "genetic markers", "probability of paternity", "likelihood ratio",
        "markers", "interpretation of results", "probability margin", "signatures"
    ]

    filtered_analysis = []

    for line in content.split("\n"):
        if any(keyword in line for keyword in analysis_keywords):
            filtered_analysis.append(line)

    return "\n".join(filtered_analysis)

def display_analysis(extracted_text):
    """
    Display the main analysis of the DNA report.
    """
    # Get the filtered DNA analysis
    dna_analysis = parse_dna_analysis(extracted_text)

    if dna_analysis:
        st.write("### DNA Analysis:")
        st.text_area("Main Analysis", dna_analysis, height=300)
    else:
        st.warning("No DNA analysis information found in the content.")

# Streamlit app
st.title("DNA Image Analysis with Groq")

# File uploader widget
uploaded_file = st.file_uploader("Upload a DNA Image", type=["png", "jpg", "jpeg"])

if uploaded_file is not None:
    # Open the uploaded image
    image = Image.open(uploaded_file)
    st.image(image, caption="Uploaded Image", use_container_width=True)

    # Save the image to a temporary file for Groq analysis
    image_path = "/content/temp_uploaded_image.jpg"
    image.save(image_path)

    # Perform analysis using Groq
    extracted_text = analyze_image_with_groq(image_path)

    if extracted_text:
        # Filter and display only the main DNA analysis part
        display_analysis(extracted_text)
    else:
        st.warning("No analysis result received from the API.")
else:
    st.info("Please upload an image to extract text and analyze using Groq.")
```
