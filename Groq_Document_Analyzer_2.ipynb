```python
def extract_text_from_pdf(pdf_path):
    try:
        text = ""
        # Read PDF file
        reader = PdfReader(pdf_path)
        for page in reader.pages:
            text += page.extract_text()
        return text
    except Exception as e:
        print(f"Error extracting text from PDF: {e}")
        return ""
```

```python
def extract_text_from_images(pdf_path):
    try:
        # Convert PDF to images
        images = convert_from_path(pdf_path)
        text = ""
        for img in images:
            text += image_to_string(img)
        return text
    except Exception as e:
        print(f"Error extracting text from images: {e}")
        return ""
```

```python
def process_pdf_with_groq(pdf_path):
    # Extract text directly from PDF
    text = extract_text_from_pdf(pdf_path)

    # If text extraction fails, use OCR on images
    if not text.strip():
        print("PDF text extraction failed. Using OCR...")
        text = extract_text_from_images(pdf_path)

    if text.strip():
        print("Sending text to Groq API for analysis...")
        result = analyze_text_with_groq(text)
        return result
    else:
        print("No text found in the document.")
        return None
```

```python
import os
from PyPDF2 import PdfReader
from pdf2image import convert_from_path
import pytesseract
import requests
```

```python
# Replace these with actual values
GROQ_API_URL = "https://console.groq.com/playground?model=llama-3.2-90b-vision-preview"  # Replace with the actual Groq API URL
GROQ_API_KEY = "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"  # Replace with your Groq API key
```

```python
def extract_text_from_pdf(pdf_path):
    """
    Extract text from a PDF file using PyPDF2 and fallback to OCR.
    """
    try:
        # First, try extracting text directly from the PDF
        reader = PdfReader(pdf_path)
        text = ""
        for page in reader.pages:
            text += page.extract_text()
        if text.strip():
            return text
        else:
            print("PDF text extraction failed. Using OCR...")
    except Exception as e:
        print(f"Error reading PDF: {e}")
        print("Using OCR as a fallback...")

    # Fallback to OCR
    try:
        images = convert_from_path(pdf_path)
        text = ""
        for img in images:
            text += pytesseract.image_to_string(img)
        return text if text.strip() else None
    except Exception as e:
        print(f"Error extracting text from images: {e}")
        return None
```

```python
def process_pdf_with_groq(pdf_path):
    """
    Process the PDF file using the Groq API.
    """
    # Step 1: Extract text from the PDF
    extracted_text = extract_text_from_pdf(pdf_path)
    if not extracted_text:
        print("No text found in the document.")
        return None

    # Step 2: Send the extracted text to the Groq API
    try:
        headers = {
            "Authorization": f"Bearer {GROQ_API_KEY}",
            "Content-Type": "application/json"
        }
        payload = {
            "model": "llama-3.2-90b-vision-preview",  # Specify the model
            "prompt": extracted_text
        }
        response = requests.post(GROQ_API_URL, headers=headers, json=payload)

        if response.status_code == 200:
            return response.json()  # Parse the response JSON
        else:
            print(f"Error from Groq API: {response.status_code} - {response.text}")
            return None
    except Exception as e:
        print(f"Error connecting to Groq API: {e}")
        return None
```
